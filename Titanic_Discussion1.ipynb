```python
import seaborn as sns
sns.set(style="ticks", palette="bright")
sns.pairplot(x_vars=["Age"], y_vars=["Survived"], data=dfTitanic, hue="Sex",  
             palette=["g", "b"],height = 6)
```

```python
sns.set(style="ticks", palette="bright")
sns.pairplot(x_vars=["Age"], y_vars=["Survived"], data=dfTitanic, hue="Pclass",  
             palette=["g", "b", "r"],height = 6)
```

```python
sns.set(style="ticks", palette="bright")
sns.pairplot(x_vars=["Age"], y_vars=["Fare"], data=dfTitanic, hue="Sex",  
             palette=["g", "b"],height = 6)
```

```python
sns.set(style="ticks", palette="bright")
sns.pairplot(x_vars=["Age"], y_vars=["Fare"], data=dfTitanic, hue="Survived",  
             palette=["g", "b"],height = 6)
```
